```python
# 6. RGB to CMYK (CORREGIDO - Problema de indentación solucionado)
print("\n6. Convirtiendo RGB a CMYK...")
try:
    img_cmyk = io.imread('images/images/colores.png')
    rgb = img_cmyk.copy()
    rgb_p = rgb.astype('float32') / 255  # Cambiado a float32
    
    with np.errstate(invalid='ignore', divide='ignore'):
        K = 1 - np.max(rgb_p, axis=2)  # INDENTADO CORRECTAMENTE
        # Extrae los canales de acuerdo con la ecuación
        C = (1 - rgb_p[:,:,0] - K) / (1 - K + 1e-10)  # Evitar división por cero
        M = (1 - rgb_p[:,:,1] - K) / (1 - K + 1e-10)
        Y = (1 - rgb_p[:,:,2] - K) / (1 - K + 1e-10)
    
    # Limpiar valores NaN
    C = np.nan_to_num(C, nan=0.0, posinf=1.0)
    M = np.nan_to_num(M, nan=0.0, posinf=1.0)
    Y = np.nan_to_num(Y, nan=0.0, posinf=1.0)
    K = np.nan_to_num(K, nan=0.0, posinf=1.0)
    
    CMYK = (np.dstack((C, M, Y, K)) * 255).astype('uint8')
    
    # Extraer canales
    C_channel = CMYK[:,:,0]
    M_channel = CMYK[:,:,1]
    Y_channel = CMYK[:,:,2]
    K_channel = CMYK[:,:,3]
    
    fig, ax = plt.subplots(2, 3, figsize=(12, 8))
    ax[0,0].imshow(img_cmyk)
    ax[0,0].set_title('Original RGB')
    ax[0,0].axis('off')
    
    ax[0,1].imshow(CMYK[:,:,:3])  # Mostrar como RGB
    ax[0,1].set_title('CMYK (visual)')
    ax[0,1].axis('off')
    
    ax[0,2].imshow(C_channel, cmap='gray')
    ax[0,2].set_title('Cyan')
    ax[0,2].axis('off')
    
    ax[1,0].imshow(M_channel, cmap='gray')
    ax[1,0].set_title('Magenta')
    ax[1,0].axis('off')
    
    ax[1,1].imshow(Y_channel, cmap='gray')
    ax[1,1].set_title('Yellow')
    ax[1,1].axis('off')
    
    ax[1,2].imshow(K_channel, cmap='gray')
    ax[1,2].set_title('Black')
    ax[1,2].axis('off')
    
    plt.suptitle('Conversión RGB a CMYK')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"No se pudo cargar la imagen 'colores.png': {e}")
```

```python
# 7. OTRAS CONVERSIONES
print("\n7. Realizando otras conversiones de color...")
img_otras = cv2.imread('images/images/Lena_RGB.png')
if img_otras is not None:
    gray_img = cv2.cvtColor(img_otras, cv2.COLOR_BGR2GRAY)
    HSV = cv2.cvtColor(img_otras, cv2.COLOR_BGR2HSV)
    Lab = cv2.cvtColor(img_otras, cv2.COLOR_BGR2Lab)
    YCrCb = cv2.cvtColor(img_otras, cv2.COLOR_BGR2YCrCb)
    
    fig, ax = plt.subplots(2, 2, figsize=(10, 10))
    
    # Grayscale - CORRECTO
    ax[0,0].imshow(gray_img, cmap='gray')
    ax[0,0].set_title('Grayscale')
    ax[0,0].axis('off')
    
    # HSV - CORRECTO (así se ve el color real)
    ax[0,1].imshow(cv2.cvtColor(HSV, cv2.COLOR_HSV2RGB))
    ax[0,1].set_title('HSV')
    ax[0,1].axis('off')
    
    # LAB - CORREGIDO: Mostrar canales individualmente
    # Normalizamos cada canal para poder verlo
    L_norm = (Lab[:,:,0] - Lab[:,:,0].min()) / (Lab[:,:,0].max() - Lab[:,:,0].min())
    a_norm = (Lab[:,:,1] - Lab[:,:,1].min()) / (Lab[:,:,1].max() - Lab[:,:,1].min())
    b_norm = (Lab[:,:,2] - Lab[:,:,2].min()) / (Lab[:,:,2].max() - Lab[:,:,2].min())
    
    lab_display = np.stack([L_norm, a_norm, b_norm], axis=2)
    ax[1,0].imshow(lab_display)
    ax[1,0].set_title('LAB (canales normalizados)')
    ax[1,0].axis('off')
    
    # YCrCb - CORREGIDO: Mostrar canales individualmente
    Y_norm = (YCrCb[:,:,0] - YCrCb[:,:,0].min()) / (YCrCb[:,:,0].max() - YCrCb[:,:,0].min())
    Cr_norm = (YCrCb[:,:,1] - YCrCb[:,:,1].min()) / (YCrCb[:,:,1].max() - YCrCb[:,:,1].min())
    Cb_norm = (YCrCb[:,:,2] - YCrCb[:,:,2].min()) / (YCrCb[:,:,2].max() - YCrCb[:,:,2].min())
    
    ycrcb_display = np.stack([Y_norm, Cr_norm, Cb_norm], axis=2)
    ax[1,1].imshow(ycrcb_display)
    ax[1,1].set_title('YCrCb (canales normalizados)')
    ax[1,1].axis('off')
    
    plt.suptitle('Diferentes Espacios de Color')
    plt.tight_layout()
    plt.show()
```

# TRANSFORMACIONES GEOMETRICAS

```python
# 8. TRASLACIÓN
print("\n8. Aplicando traslación...")
try:
    img_got = io.imread('images/images/GOT.png')
    rows, cols, ch = img_got.shape
    M = np.float32([[1,0,200],[0,1,-60]])
    new_img = cv2.warpAffine(img_got, M, (cols, rows))
    
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    ax[0].imshow(img_got)
    ax[0].set_title('Original')
    ax[0].axis('off')
    ax[1].imshow(new_img)
    ax[1].set_title('Traslación (dx=200, dy=-60)')
    ax[1].axis('off')
    plt.suptitle('Transformación de Traslación')
    plt.show()
except Exception as e:
    print(f"No se pudo cargar 'GOT.png': {e}")
```

```python
# 9. CROPPING
print("\n9. Aplicando cropping...")
try:
    img_crop = io.imread('images/images/GOT.png')
    new_img_crop = img_crop[90:290, 200:305]
    
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    ax[0].imshow(img_crop)
    ax[0].set_title('Original')
    ax[0].axis('off')
    ax[1].imshow(new_img_crop)
    ax[1].set_title('Cropping [90:290, 200:305]')
    ax[1].axis('off')
    plt.suptitle('Recorte de Imagen')
    plt.show()
except Exception as e:
    print(f"Error en cropping: {e}")
```

```python
# 10. ROTACIÓN
print("\n10. Aplicando rotación...")
try:
    img_rot = io.imread('images/images/GOT.png')
    rows, cols, ch = img_rot.shape
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle=90, scale=2)
    new_img_rot = cv2.warpAffine(img_rot, M, (cols, rows))
    
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    ax[0].imshow(img_rot)
    ax[0].set_title('Original')
    ax[0].axis('off')
    ax[1].imshow(new_img_rot)
    ax[1].set_title('Rotación 90° + Escala 2x')
    ax[1].axis('off')
    plt.suptitle('Rotación y Escalamiento')
    plt.show()
except Exception as e:
    print(f"Error en rotación: {e}")
```

```python
# 11. PERSPECTIVA
print("\n11. Aplicando transformación de perspectiva...")
try:
    img_pers = cv2.imread('images/images/sudoku.png')
    if img_pers is not None:
        rows, cols, ch = img_pers.shape
        pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
        pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
        
        plt.figure(figsize=(8,6))
        plt.imshow(cv2.cvtColor(img_pers, cv2.COLOR_BGR2RGB))
        for i in range(0,4):
            plt.plot(pts1[i,0], pts1[i,1], 'or', markersize=10, label='Puntos origen' if i==0 else '')
            plt.plot(pts2[i,0], pts2[i,1], 'ob', markersize=10, label='Puntos destino' if i==0 else '')
        plt.title('Keypoints para perspectiva')
        plt.legend()
        plt.show()
        
        M = cv2.getPerspectiveTransform(pts1, pts2)
        pers = cv2.warpPerspective(img_pers, M, (305, 305))
        crop = img_pers[50:400, 20:400]
        
        fig, ax = plt.subplots(1,2, figsize=(12,6))
        ax[0].imshow(cv2.cvtColor(pers, cv2.COLOR_BGR2RGB))
        ax[0].set_title('Perspectiva Corregida')
        ax[0].axis('off')
        ax[1].imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
        ax[1].set_title('Cropping')
        ax[1].axis('off')
        plt.suptitle('Transformación de Perspectiva')
        plt.show()
    else:
        print("No se encontró 'sudoku.png'")
except Exception as e:
    print(f"Error en perspectiva: {e}")
```

```python
# 12. FLIPPING
print("\n12. Aplicando flipping...")
try:
    img_flip = io.imread('images/images/Lena_RGB.png')
    flipVertical = cv2.flip(img_flip, 0)
    flipHorizontal = cv2.flip(img_flip, 1)
    flipBoth = cv2.flip(img_flip, -1)
    
    fig, ax = plt.subplots(1,4, figsize=(16,4))
    ax[0].imshow(img_flip)
    ax[0].set_title('Original')
    ax[0].axis('off')
    ax[1].imshow(flipVertical)
    ax[1].set_title('Flip Vertical')
    ax[1].axis('off')
    ax[2].imshow(flipHorizontal)
    ax[2].set_title('Flip Horizontal')
    ax[2].axis('off')
    ax[3].imshow(flipBoth)
    ax[3].set_title('Flip Both')
    ax[3].axis('off')
    plt.suptitle('Transformaciones de Volteo (Flipping)')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error en flipping: {e}")

print("\n=== FIN DEL PROCESAMIENTO ===")
```
